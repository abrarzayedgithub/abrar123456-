```python
#using filter2d
image = cv2.imread('/content/drive/MyDrive/iiuc mosque.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

kernel1 = np.ones((3,3),np.float32)/9
blur_img1 = cv2.filter2D(image,-1,kernel1)

kernel2 = np.ones((5,5),np.float32)/25
blur_img2 = cv2.filter2D(image,-1,kernel2)



plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.title('original image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1,3,2)
plt.title('averaging 3*3')
plt.imshow(blur_img1)
plt.axis('off')

plt.subplot(1,3,3)
plt.title('averaging 5*5')
plt.imshow(blur_img2)
plt.axis('off')

plt.show()
```

```python
#using cv2.blur(image,(k,k))
image = cv2.imread('/content/drive/MyDrive/house.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

blur_img1 = cv2.blur(image,(3,3))
blur_img2 = cv2.blur(image,(5,5))

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.title('original image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1,3,2)
plt.title('averaging 3*3')
plt.imshow(blur_img1)
plt.axis('off')

plt.subplot(1,3,3)
plt.title('averaging 5*5')
plt.imshow(blur_img2)
plt.axis('off')

plt.show()
```

```python
#other blur like avg blur,gaussian blur,bilateral blur,median blur
image = cv2.imread('/content/drive/MyDrive/house.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

avg_blur = cv2.blur(image,(3,3))
gaussian_blur = cv2.GaussianBlur(image,(5,5),0)
bilateral_blur = cv2.bilateralFilter(image,5,75,75)
median_blur = cv2.medianBlur(image,3)

plt.figure(figsize=(12,6))
plt.subplot(1,4,1)
plt.title('average')
plt.imshow(avg_blur,cmap='gray')
plt.axis('off')

plt.subplot(1,4,2)
plt.title('gaussian')
plt.imshow(gaussian_blur,cmap='gray')
plt.axis('off')

plt.subplot(1,4,3)
plt.title('bilateral_blur')
plt.imshow(bilateral_blur,cmap='gray')
plt.axis('off')

plt.subplot(1,4,4)
plt.title('median_blur')
plt.imshow(median_blur,cmap='gray')
plt.axis('off')

plt.show()
```
