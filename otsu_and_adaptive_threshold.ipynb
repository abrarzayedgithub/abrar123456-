```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#load image
image_path = '/content/drive/MyDrive/berry-1.jpg'
image = plt.imread(image_path)
plt.imshow(image)
print(image.shape)

#convert to gary scale
gray_scale_image  = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#apply global thresholding
ret1,th1 = cv2.threshold(gray_scale_image,60,255,cv2.THRESH_BINARY)
#apply otsu's thrersholding

ret2,th2 = cv2.threshold(gray_scale_image,60,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

#apply adaptive mean thresholding
th3 = cv2.adaptiveThreshold(gray_scale_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,7,2)

#apply adaptive gaussian thresholding
th4 = cv2.adaptiveThreshold(gray_scale_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,2)

#plot the result
titles = ['original_imge','Global Thresholding (v=127)',"otsu's threshoding",'adaptive gaussian thresholding']
images = [gray_scale_image,th1,th2,th4]

for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i],cmap = 'gray')
  plt.title(titles[i])
  plt.xticks([])
  plt.yticks([])
plt.show()

```
