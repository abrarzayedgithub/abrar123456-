HW-1: separate the orange from given image (single color)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/orange-1 (1).jpg')
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

#define range the blue color of hsv
lower_yellow = np.array([15,100,100])
upper_yellow = np.array([30,255,255])

th = cv2.inRange(image_hsv,lower_yellow,upper_yellow)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('original image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(th,cmap='gray')
plt.title('threshold image') # Corrected title
plt.axis('off')

plt.show()

```

HW-3: build mask(circle) and apply on the given image so that only the orange will be shown

```python
#get the dimension of the image
height,width,_ = image.shape
mask = np.zeros((height,width),dtype=np.uint8)

#define the circle center and the radius
center = (width//2,height//2)
radius = min(width,height)//4

cv2.circle(mask,center,radius,(255),thickness = -1)

#convert the mask to the 3 channel
mask_rgb = cv2.merge([mask,mask,mask])

#muktiply the original image with the mask
result = cv2.bitwise_and(image,mask_rgb)

#convert the original image and result from BGR to RGB
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
result_rgb = cv2.cvtColor(result,cv2.COLOR_BGR2RGB)

#display the image using matplotlib
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('original image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(result_rgb)
plt.title('mask image')
plt.axis('off')

plt.show()
```
