```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#load image
image = cv2.imread('/content/drive/MyDrive/iiuc mosque.jpg',cv2.IMREAD_GRAYSCALE)

#noise reduce using gaussian blur
noise_reduce = cv2.GaussianBlur(image,(5,5),0)

#adaptive gaussian thresholding
thresh1 = cv2.adaptiveThreshold(noise_reduce,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,3)

titles = ['original','blurred image','thresholded image']
images = [image,noise_reduce,thresh1]


for i in range(3):
  plt.subplot(1,3,i+1)
  plt.imshow(images[i],'gray',vmin=0,vmax=255)
  plt.title(titles[i])
  plt.xticks([])
  plt.yticks([])
plt.show()
```
