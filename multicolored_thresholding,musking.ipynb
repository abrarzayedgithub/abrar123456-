```python
image = cv2.imread('/content/drive/MyDrive/iiuc mosque.jpg')
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

#define range of blue color in hsv
lower_yellow = np.array([20,100,100])
upper_yellow = np.array([30,255,255])

thr = cv2.inRange(image_hsv,lower_yellow,upper_yellow)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(image_rgb)
plt.axis('off')


plt.subplot(1,2,2)
plt.title('mask')
plt.imshow(thr,cmap='gray')
plt.axis('off')

plt.show()
```

```python
image = cv2.imread('/content/drive/MyDrive/iiuc mosque.jpg')
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#get the dimension of the image
height,width,_ = image.shape

mask = np.zeros((height,width,3),dtype = np.uint8)
square_size = 400
top_left_x = (width - square_size)//2
top_left_y = (height - square_size)//2

#create a white square in the mask
mask[top_left_y:top_left_y + square_size, top_left_x:top_left_x + square_size] = 255
plt.imshow(mask)
```

```python
result = cv2.bitwise_and(image,mask)
plt.imshow(result)
```
