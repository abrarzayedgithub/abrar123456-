```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#load the image
image_path ='/content/drive/MyDrive/berry-1.jpg'
image = plt.imread(image_path)
plt.imshow(image)
print(image.shape)
#convert to gray scale image
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#apply otsu
ret,th = cv2.threshold(gray_image,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.subplot(1,3,1)
plt.title("image")
plt.imshow(image,cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("image")
plt.imshow(gray_image,cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title(f"otsu's Threshloding (T = {int(ret)})")
plt.imshow(th,cmap='gray')
plt.axis('off')
plt.show()
```
